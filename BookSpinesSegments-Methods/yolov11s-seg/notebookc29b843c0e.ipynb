```python
from ultralytics import YOLO
```

```python
model = YOLO(r"/kaggle/input/yolo11s-seg/pytorch/default/1/yolo11s-seg.pt")
```

```python
model.train(data="/kaggle/input/bookspinedataset/data.yaml", imgsz=640, device=0, batch=16, epochs=300, workers=4, project='/kaggle/working/', name='yolov11s-seg_training')
```

```python
import shutil

# Save all files from 'kaggle/working' to 'kaggle/outputs' when training completes
shutil.make_archive('/kaggle/outputs/training_results', 'zip', '/kaggle/working')
```
